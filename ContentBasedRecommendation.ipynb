```python
import ast
```

```python
def convert(obj):
    L =[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
df['genres'] = df['genres'].apply(convert)
df['keywords'] = df['keywords'].apply(convert)
```

```python
df.head()
```

```python
def convert2(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter =+ 1
        else:
            break
        return L
```

```python
df['cast'] = df['cast'].apply(convert2)
df.head()
```

```python
df['crew'] = df['crew'].apply(convert2)
df.head()
```

```python
df['overview'] = df['overview'].apply(lambda x: x.split())
```

```python
df = df.dropna(subset=['keywords'])  # Drop rows where 'cast' is None before applying the lambda function
df['keywords'] = df['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
def stem(text_list):
    if isinstance(text_list, list):
        stemmed_words = []
        for text in text_list:
            if isinstance(text, str):  # Check if it's a string
                stemmed_text = " ".join([ps.stem(word) for word in text.split()])
                stemmed_words.append(stemmed_text)
            else:
                stemmed_words.append("")  # Handle non-string values
        return stemmed_words
    else:
        return [""] if pd.isnull(text_list) else [ps.stem(str(text_list))]  # Handle NaN or non-list values
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def visualize_recommendations(movie, recommended_movies):
    G = nx.Graph()

    # Add input movie node to the graph
    G.add_node(movie)

    # Add recommended movies as nodes and connect them to the input movie
    for rec_movie in recommended_movies:
        G.add_node(rec_movie)
        G.add_edge(movie, rec_movie)

    # Draw the network graph
    plt.figure(figsize=(10, 8))
    pos = nx.spring_layout(G, k=0.5)
    nx.draw(G, pos, with_labels=True, font_weight='bold')
    plt.title(f'Recommended Movies for {movie}')
    plt.show()

def recommend2(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    recommended_movies = [new_df.iloc[i[0]]['title'] for i in movie_list]
    visualize_recommendations(movie, recommended_movies)



```
